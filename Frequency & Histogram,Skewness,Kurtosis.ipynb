```python
frequency_table=pd.DataFrame()
```

```python
import numpy as np
```

```python
def freq_table(dataset,columnName):
    frequency_table=pd.DataFrame(columns=["Unique_Values","Frequency","Relative_Frequency","Cumulative_Frequency"])
    unique_count=dataset[columnName].nunique()
    frequency_table["Unique_Values"]=dataset[columnName].value_counts().index
    frequency_table["Frequency"]=dataset[columnName].value_counts().values
    frequency_table["Relative_Frequency"]=frequency_table["Frequency"]/unique_count
    frequency_table["Cumulative_Frequency"]=dataset[columnName].cumsum()
    return frequency_table
```

```python
freq_table(dataset,"salary")
```
